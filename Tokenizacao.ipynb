```python
#Vocabulário
lista_palavras_feliz = ["alegre","felicidade","contente","legal","massa"]
lista_palavras_triste = ["tristeza","chorar","magoado","saudade","triste"]

tende_feliz = 0
tende_triste = 0
```

```python
#Tratamento da entrada da frase
frase = "A tristeza me faz chorar mesmo quando estou magoado."

frase_tratada = frase.lower().replace(".","")
palavras = frase_tratada.split()

print(palavras)

for token in palavras:

  if token in lista_palavras_feliz:
    tende_feliz += 1
  elif token in lista_palavras_triste:
    tende_triste += 1

if tende_triste > tende_feliz:
  print("Triste")
elif tende_feliz > tende_triste:
  print("Feliz")




```
