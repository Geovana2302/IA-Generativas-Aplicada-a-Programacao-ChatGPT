## **Exercício 3: Faça uma pesquisa sobre as bibliotecas statistics e numpy.**


### Comparativo: Biblioteca `statistics` vs. `numpy`

Se você está começando a manipular dados com Python, é muito provável que encontre essas duas ferramentas. Embora ambas lidem com números, elas ocupam espaços bem diferentes no ecossistema da linguagem.

---

## 1. A Biblioteca `statistics`
Lançada no Python 3.4, a `statistics` faz parte da **Standard Library**. Isso significa que você não precisa instalar nada; ela já vem "na caixa" com o Python.

* **Foco:** Cálculos estatísticos básicos em dados de tamanho pequeno a moderado.
* **Público:** Desenvolvedores que precisam de métricas rápidas sem dependências externas.
* **Vantagem:** Simplicidade extrema e integração nativa com listas e tuplas.

### Principais Funções
* `mean()`: Média aritmética.
* `median()`: Mediana.
* `mode()`: Moda.
* `stdev()`: Desvio padrão amostral.

---

## 2. A Biblioteca `numpy`
O **NumPy** (Numerical Python) é a base da computação científica em Python. É uma biblioteca de terceiros (`pip install numpy`).

* **Foco:** Processamento de arrays multidimensionais de alto desempenho.
* **Público:** Cientistas de dados e engenheiros lidando com grandes volumes de dados.
* **Vantagem:** Velocidade. O NumPy é escrito em C, sendo muito mais rápido que as listas nativas.

O NumPy introduz o **ndarray**, que permite cálculos vetoriais (operar em toda a lista de uma vez, sem `for loops`).

---

## Comparativo Direto

| Característica | `statistics` | `numpy` |
| :--- | :--- | :--- |
| **Instalação** | Nativa | Externa (`pip install numpy`) |
| **Performance** | Lenta para grandes volumes | Extremamente rápida (Baseada em C) |
| **Tipos de Dados** | Listas padrão | Arrays homogêneos (`ndarray`) |
| **Complexidade** | Simples | Curva de aprendizado maior |
| **Recursos** | Estatística básica | Álgebra linear, Fourier, etc. |

---

## Quando usar cada uma?

### Use `statistics` se:
* Seu script deve ser simples (sem arquivos de dependências).
* Seus dados são pequenos.
* Você precisa apenas do básico (ex: média de notas).

### Use `numpy` se:
* Você trabalha com Data Science ou Machine Learning.
* Tem milhões de pontos de dados.
* Precisa de fórmulas complexas ou álgebra linear:

$$\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}$$

## **Exercício 4: Calcular a mediana utilizando a biblioteca statistics, explicar cada linha de código.**

## **Exercício 5: Faça uma pesquisa sobre desvio padrão.**


### O que é Desvio Padrão?

Enquanto a **Média** nos dá o valor central dos dados, o **Desvio Padrão** nos diz o quão *confiável* essa média é. Ele mede a **dispersão**: o quanto os seus dados variam em relação à média.

* **Desvio Padrão Baixo:** Os dados são consistentes e estão todos "agrupados" perto da média. (Ex: Temperatura de uma geladeira: 4°C, 3.9°C, 4.1°C).
* **Desvio Padrão Alto:** Os dados são voláteis e espalhados. (Ex: Temperatura de uma cidade no deserto: 0°C a noite, 40°C de dia).

---

## 1. A Matemática da Coisa (Fórmulas)

Existem duas formas de calcular, dependendo do que seus dados representam. A diferença está no divisor (N vs N-1).

### A. Desvio Padrão Populacional ($\sigma$)
Use quando você tem **todos** os dados do seu universo de interesse (ex: notas de *todos* os alunos da sua sala).

$$\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}$$

### B. Desvio Padrão Amostral ($s$)
Use quando você tem apenas uma **amostra** de um grupo maior (ex: pesquisa de opinião com 1.000 pessoas para representar o Brasil). Dividimos por $N-1$ para corrigir distorções estatísticas (Correção de Bessel).

$$s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N - 1}}$$

---

## 2. A "Pegadinha" do Python (`numpy` vs `statistics`)

Aqui mora um erro comum de Data Science.
* A biblioteca `statistics` calcula o desvio **Amostral** por padrão.
* A biblioteca `numpy` calcula o desvio **Populacional** por padrão.

Se você não ajustar isso, seus resultados serão diferentes!

## **Exercício 6: Gerar o desvio padrão de um conjunto de números, utilizando a biblioteca numpy, explicar cada linha de código.**


```python
# Importa a biblioteca numpy com o apelido 'np' para facilitar o uso
import numpy as np

# Variaveis para entrada dos dados pelo usuario
print("Vamos calcular o desvio padrão de 4 valores!")
v1 = float(input("Digite o primeiro valor: "))
v2 = float(input("Digite o segundo valor: "))
v3 = float(input("Digite o terceiro valor: "))
v4 = float(input("Digite o quarto valor: "))

# Cria um array do numpy (estrutura de alta performance) com os valores recebidos
lista_valores = np.array([v1, v2, v3, v4])

# Calculo do desvio padrão utilizando a função otimizada do numpy
desvio_padrao = np.std(lista_valores)

# Exibe o resultado do desvio padrão formatado com duas casas decimais
print(f"\nO desvio padrão dos dados é: {desvio_padrao:.2f}")
```

## **Exercício 7: Faça uma pesquisa sobre Amplitude, utilizado em Estatística, faça um exemplo utilizando uma lista de dados.**


### O que é Amplitude em Estatística?

A **Amplitude** é a diferença exata entre o maior e o menor valor de um conjunto de dados. Ela serve para mostrar a extensão total da variação dos seus dados.

### A Fórmula
A matemática aqui é a mais direta possível:

$$Amplitude = X_{max} - X_{min}$$

### Vantagens e Desvantagens
* **Vantagem:** É extremamente fácil de calcular e de explicar para qualquer pessoa.
* **Desvantagem (A Pegadinha):** Ela é **altamente sensível a *outliers*** (valores atípicos). Se você tem 99 produtos que custam entre R$10 e R$ 20, mas um único produto que custa R$ 5.000, a sua amplitude será enorme e não representará bem a realidade da maioria do seu estoque.

```python
import numpy as np

# Nossos dados: Salários de uma pequena empresa (repare no último valor)
salarios = [1500, 1600, 1550, 1700, 1800, 25000] # O chefe ganha 25.000

# --- 1. Calculando a Amplitude com Python Nativo ---
amplitude_nativa = max(salarios) - min(salarios)

# --- 2. Calculando a Amplitude com NumPy ---
# O método ptp() calcula automaticamente max - min
amplitude_numpy = np.ptp(salarios)

print(f"Lista de Salários: {salarios}")
print("-" * 40)
print(f"Maior salário: R$ {max(salarios)}")
print(f"Menor salário: R$ {min(salarios)}")
print("-" * 40)
print(f"Amplitude (Python nativo): R$ {amplitude_nativa}")
print(f"Amplitude (NumPy ptp):     R$ {amplitude_numpy}")

# Explicando o problema do Outlier
print("\nNota: A amplitude de 23.500 passa a falsa impressão de que os salários variam muito entre os funcionários comuns, quando na verdade é apenas um valor atípico distorcendo a métrica.")
```

## **Exercício 8: Realizar a análise preditiva do funcionamento de um ar condicionado em um determinado ambiente. (desvio padrão, Amplitude).**


# Exercício 8: Análise Preditiva de um Ar Condicionado

Neste cenário, vamos simular o monitoramento de temperatura de um servidor que deve permanecer estritamente a **23°C**. Usaremos estatística para diagnosticar a saúde do equipamento.

### Os Indicadores de Saúde

1.  **Média ($\mu$):** Indica se o aparelho consegue atingir a temperatura alvo.
    * *Ideal:* Próximo de 23°C.
    * *Problema:* Se a média for 26°C, o aparelho não está dando conta (subdimensionado ou filtro sujo).

2.  **Amplitude ($Range$):** A diferença entre o pico de calor e o pico de frio.
    * *Ideal:* Baixa (menor que 2°C).
    * *Problema:* Amplitude alta indica que o ambiente esquenta muito antes do ar ligar novamente.

3.  **Desvio Padrão ($\sigma$):** A métrica mais crítica para conforto e precisão.
    * *Baixo (< 0.5):* O ar é "Inverter" e muito estável.
    * *Alto (> 2.0):* O ar está oscilando demais ("gelando e parando"), o que gasta muita energia e desgasta peças.

---
### O Cenário
Vamos comparar dois aparelhos:
* **Aparelho A (Inverter):** Promete estabilidade.
* **Aparelho B (Antigo/Defeituoso):** Liga e desliga com frequência (Histerese alta).

```python
import numpy as np
import statistics

# --- 1. Gerando Dados Simulados (Temperaturas lidas a cada minuto) ---
# Aparelho A: Mantém 23°C com variação mínima (Tecnologia Inverter)
# Gera 60 leituras com média 23 e variação de 0.5
aparelho_a = np.random.normal(loc=23.0, scale=0.3, size=60)

# Aparelho B: Oscila entre 20°C e 26°C (Compressor liga/desliga ou defeito)
# Gera dados mais caóticos
aparelho_b = np.concatenate([
    np.linspace(20, 26, 15), # Esquentando
    np.linspace(26, 20, 15), # Esfriando
    np.linspace(20, 26, 15),
    np.linspace(26, 20, 15)
])

# --- 2. Função de Análise Preditiva ---
def analisar_saude(nome, dados):
    media = statistics.mean(dados)
    amplitude = np.ptp(dados) # Peak to Peak (Max - Min)
    desvio = statistics.stdev(dados)

    print(f"--- Diagnóstico: {nome} ---")
    print(f"Temperatura Média: {media:.2f}°C")
    print(f"Amplitude:         {amplitude:.2f}°C")
    print(f"Desvio Padrão:     {desvio:.2f}°C")

    # Regras de Negócio (Preditiva)
    if desvio > 1.5:
        print("STATUS: ⚠️ CRÍTICO - Oscilação excessiva. Risco de falha no termostato ou sensor.")
    elif abs(media - 23) > 2:
        print("STATUS: ⚠️ ALERTA - Não atinge o alvo (23°C). Verificar filtros ou gás.")
    else:
        print("STATUS: ✅ NORMAL - Operação eficiente.")
    print("\n")

# --- 3. Executando a Análise ---
analisar_saude("Ar Condicionado A (Novo)", aparelho_a)
analisar_saude("Ar Condicionado B (Velho)", aparelho_b)
```
