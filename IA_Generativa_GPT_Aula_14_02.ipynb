```python
# Operações Aritimeticas



num_op = int(input("Digite a operação que deseja\n 1-Soma\n 2-Substração\n 3-Multiplicação\n 4-Exponenciação\n 5-Divisão\n 6-DivInteira\n 7-RestoDiv\n "))
num_1 = float(input("Digite o primeiro numero: "))
num_2 = float(input("Digite o segundo numero: "))
if num_op == 1:
  resultado = num_1 + num_2
  print(f"Resultado da Soma: {resultado}")
elif num_op == 2:
  resultado = num_1 - num_2
  print(f"Resultado da Subtração: {resultado}")
elif num_op == 3:
  resultado = num_1 * num_2
  print(f"Resultado da Multiplicação: {resultado}")
elif num_op == 4:
  resultado = num_1 ** num_2
  print(f"Resultado da Exponenciação: {resultado}")
elif num_op == 5:
  resultado = num_1 / num_2
  print(f"Resultado da Divisão: {resultado}")
elif num_op == 6:
  resultado = num_1 // num_2
  print(f"Resultado da Divisão Inteira: {resultado}")
elif num_op == 7:
  resultado = num_1 % num_2
  print(f"Resultado do resto da divisão: {resultado}")
elif num_op >=8:
  print("Não tem operações com esse numero")
```
