```python
# Baixando o modelo de inglês do SpaCy
!python -m spacy download en_core_web_sm

import spacy
from textblob import TextBlob

# Carregando o modelo para a memória
nlp = spacy.load("en_core_web_sm")

print("Tudo pronto para começar a análise!")
```

```python
def classificar_sentimento(texto):
    # 1. Passamos o texto pelo SpaCy (NLP)
    doc = nlp(texto)

    # 2. O TextBlob analisa o texto processado
    analise = TextBlob(doc.text)
    polaridade = analise.sentiment.polarity

    # 3. Definimos os limites (Positive, Negative, Neutral)
    if polaridade > 0.1:
        sentimento = "Positivo"
    elif polaridade < -0.1:
        sentimento = "Negativo"
    else:
        sentimento = "Neutro"

    return sentimento, polaridade

print("Função de análise criada com sucesso! ✅")
```
