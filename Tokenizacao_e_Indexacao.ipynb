```python
vocabulario = ["dia","bom","ruim","muito"]

prompt = "muito ruim"

tokens = []
for palavra in prompt.split():
  if palavra in vocabulario:
    posicao = vocabulario.index(palavra)
    tokens.append(posicao)

print(tokens)
```

```python
pesos = [5, 10, -10, 2]

foco = "muito"
contexto = "ruim"

score = pesos[vocabulario.index(foco)] * pesos[vocabulario.index(contexto)]

print(f"Força da conexão é {score}")
```

```python
forca_total = 0

for t in tokens:
  forca_total += pesos[t]

#print(forca_total)
if forca_total > 0:
  print("Positivo")
else:
  print("Negativo")
```

#🧪 Laboratório de IA: Entendendo Embeddings e Distância

##Célula 1: Configuração do Vocabulário (Embeddings)
Aqui definimos que cada palavra é um ponto em um gráfico 2D: [Sentimento, Intensidade].


```python
import numpy as np

# Nosso "Cérebro" de IA (Dicionário de Vetores)
# [Sentimento (-1 a 1), Intensidade (0 a 2)]
embeddings = {
    "excelente": np.array([0.9, 1.8]),
    "bom":       np.array([0.7, 1.0]),
    "ruim":      np.array([-0.7, 1.0]),
    "pessimo":   np.array([-0.9, 1.8]),
    "comida":    np.array([0.1, 0.2]),
}

def calcular_similaridade(v1, v2):
    # Cálculo da Similaridade de Cosseno: mede o ângulo entre os vetores
    # Se o resultado for 1.0, são idênticos. Se for -1.0, são opostos.
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

print("✅ Vocabulário carregado!")
```

##Célula 2: O Teste de "Intuição" da IA
Agora, vamos ver se a IA entende que "excelente" é mais parecido com "bom" do que com "péssimo".



```python
# Escolha as palavras para testar
palavra_base = "excelente"
comparacoes = ["bom", "pessimo", "comida"]

print(f"Analisando a palavra: '{palavra_base}'\n" + "-"*30)

for p in comparacoes:
    sim = calcular_similaridade(embeddings[palavra_base], embeddings[p])
    print(f"Similaridade com '{p}': {sim:.4f}")

    if sim > 0.8:
        print("-> Resultado: São muito parecidas! (Sinônimos/Mesmo contexto)")
    elif sim < 0:
        print("-> Resultado: São opostas! (Antônimos)")
    else:
        print("-> Resultado: São palavras neutras entre si.")
    print("-" * 10)
```

##O que este teste explica?
* O Espaço Vetorial: As palavras não são mais apenas letras, são
coordenadas. Imagine que "bom" e "excelente" apontam para a mesma direção no mapa, enquanto "ruim" aponta para o lado contrário.

* Similaridade de Cosseno: Note que mesmo que "excelente" tenha números maiores que "bom", o ângulo entre elas é pequeno, por isso a similaridade é alta.

* Contexto: Palavras como "comida" (que são neutras no nosso modelo) terão um score baixo de similaridade com palavras de forte sentimento, pois estão em "lugares" diferentes do gráfico.
